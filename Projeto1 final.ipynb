```python
dados= pd.read_csv('DOM2009.csv',sep=',')
```

```python
dadosID = dados.loc[ : ,["UF","V0222","V0221","V0228","V4622","V4623","V2103","V2105","V2107","V2109","V4105","V4106"]]
```

```python
dadosCompletos=dadosID[pd.notnull(dadosID.V0222)]
```

```python
#dadosCompletos=dadosID.dropna()
dadosCompletos.isnull().any()
```

```python
dadosCompletos[dadosCompletos.V4106>0]
```

```python
dadosCompletos[(dadosCompletos.UF>43)&(dadosCompletos.UF<54)] #restringe a regiao centro-oeste
```

```python
dadosCompletos.to_csv("dadosp1.csv",sep=";")
          
```

```python
dadosCompletos.V4622=dadosCompletos.V4622.astype("category")
dadosCompletos.V4622.cat.categories = ("Sem Rendimento","Até 0,25SM","Entre 0,25SM e 0,5SM","Entre 0,5SM e 1SM","Entre 1SM e 2SM","Entre 2SM e 3SM","Entre 3SM e 5SM","Mais de 5SM","Sem Declaração")
dadosCompletos.V2103=dadosCompletos.V2103.astype("category")
dadosCompletos.V2105=dadosCompletos.V2105.astype("category")
dadosCompletos.V2107=dadosCompletos.V2107.astype("category")
dadosCompletos.V2109=dadosCompletos.V2109.astype("category")
dadosCompletos.loc[(dadosCompletos.V2103==1) | (dadosCompletos.V2105==1) | (dadosCompletos.V2107==1) | (dadosCompletos.V2109==1),'Insegurança_Alimentar'] = 'Sim'
dadosCompletos.loc[(dadosCompletos.V2103==3) | (dadosCompletos.V2105==3) | (dadosCompletos.V2107==3) | (dadosCompletos.V2109==3),'Insegurança_Alimentar'] = 'Não'

dadosCompletos.V2103 = dadosCompletos.V2103.astype("category") #Os moradores tiveram a preocupação de que os alimentos acabassem antes de poderem comprar ou receber mais comida
dadosCompletos.V2103.cat.categories = (['Sim','Não'])
dadosCompletos.V2107 = dadosCompletos.V2107.astype("category") #Os moradores ficaram sem dinheiro para ter uma alimentação saudável e variad
dadosCompletos.V2107.cat.categories = (['Sim','Não'])

dadosCompletos.head()
```

```python
dadosCompletos.loc[(dadosCompletos.V4105<4),"ZONA"]="Urbana"
dadosCompletos.loc[(dadosCompletos.V4105>3),"ZONA"]="Rural"
dadosCompletos.ZONA=dadosCompletos.ZONA.astype("category")

dadosCompletos.head()
```

```python
insegurança_alimentar_zona= (pd.crosstab(dadosCompletos.ZONA,dadosCompletos.Insegurança_Alimentar,margins=True,rownames=["ZONA"],colnames=["Insegurança Alimentar"],normalize='index')*100).round(decimals=1)
plot = insegurança_alimentar_zona.plot(kind="bar" ,color=('yellow', 'salmon'))
insegurança_alimentar_renda = (pd.crosstab(dadosCompletos.V4622,dadosCompletos.Insegurança_Alimentar,margins=True,rownames=["RENDA"],colnames=["Insegurança Alimentar"],normalize='index')*100).round(decimals=1)
plot = insegurança_alimentar_renda.plot(kind="bar",color=('yellow', 'salmon'))
```

```python
ct1=(pd.crosstab([dadosCompletos.V4622, dadosCompletos.ZONA],dadosCompletos.Insegurança_Alimentar, normalize='index')*100).round(decimals=1)
ct1
plot = ct1.plot(kind='bar',color=('yellow', 'salmon'))
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 2.0, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Freq.Relativa (em %)')
plt.show()
```

___
## Conclusão 

Podemos observar que a renda per capita influencia de modo diferente dependendo da zona.
Na zona urbana, como esperado quanto maior a renda per capita maior é a segurança alimentar, sem exceções.
Na zona rural observamos um comportamento diferente, em algumas exceções a segurança alimentar é menor em rendimentos per capita maiores.
